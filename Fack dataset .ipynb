```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#draw graph of any equation
x=np.linspace(1,10)
y=3*x + 9*np.sin(x) + x*np.log(x)+1/x
plt.plot(x,y)
plt.scatter(x,y,color="red")
plt.show()
```
